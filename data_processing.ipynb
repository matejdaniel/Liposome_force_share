### Read force curves 

Library jpkfile reads data and cut excessive forces. 

```python
def read_jpk(file_path: str, force_limit: float, height_cut: float) -> tuple[np.ndarray, np.ndarray]:
    """
    Reads a JPK force curve file, applies a force limit, and returns height and force data.

    This function is designed to process force spectroscopy data from JPK files. It
    extracts height and vertical deflection data, converts the deflection to a force
    value, and then truncates the data based on a specified maximum force limit.
    This is particularly useful for isolating the approach or retraction segments
    of a force curve below a certain contact force.

    Args:
        file_path (str): The full path to the JPK file.
        force_limit (float): The maximum force value to retain in the data.
                             All data points with a force greater than this
                             limit will be removed.
        height_cut (float):  The minimum height to retain in the data.
                             All data points with a height lower than this
                             limit will be removed.


    Returns:
        tuple[np.ndarray, np.ndarray]: A tuple containing two NumPy arrays:
                                        - The filtered and processed height data.
                                        - The filtered and processed force data.

    Raises:
        FileNotFoundError: If the specified file_path does not exist.
        ValueError: If the JPK file does not contain the expected 'height' or
                    'vDeflection' arrays in the first segment.
    """
    # -------------------------------------------------------------------------
    # 1. File Handling and Data Loading
    # -------------------------------------------------------------------------
    if not os.path.exists(file_path):
        raise FileNotFoundError(f"Error: The file '{file_path}' was not found.")

    try:
        jpk = jpkfile.JPKFile(file_path)
    except Exception as e:
        print(f"Failed to open JPK file: {e}")
        return None, None

    # Check for and retrieve the first segment
    if not jpk.segments:
        raise ValueError(f"JPK file '{file_path}' contains no segments.")
    
    s0 = jpk.segments[0]

    # -------------------------------------------------------------------------
    # 2. Data Extraction and Initial Processing
    # -------------------------------------------------------------------------
    # Safely retrieve the required arrays
    required_keys = ['height', 'vDeflection']
    if not all(key in s0.data for key in required_keys):
        raise ValueError(f"JPK file segment 0 is missing required data: {required_keys}")

    s0_data, _ = s0.get_array(required_keys)

    # Convert height from relative to absolute and adjust for baseline
    # JPK's 'height' is often relative, so a common practice is to invert
    # and normalize to the maximum for a sensible Z-position.
    height = s0_data['height']
    height = height.max() - height

    # Normalize force (vDeflection) data by subtracting the initial value
    # to set the baseline to zero.
    force = s0_data['vDeflection']
    force = force - force[0]

    # -------------------------------------------------------------------------
    # 3. Filtering and Final Output
    # -------------------------------------------------------------------------
    # Filter the data to include only points below the force_limit
    # This is a common practice to isolate the approach curve before contact
    # or to cut off data above a certain threshold.

    
    
    indices_to_keep = (force < force_limit) & (height > height_cut)
    
    filtered_height = height[indices_to_keep]
    filtered_force = force[indices_to_keep]

    # Squeeze to remove single-dimensional entries from the shape of an array.
    filtered_height = np.squeeze(filtered_height)
    filtered_force = np.squeeze(filtered_force)

    filtered_force = filtered_force - np.mean(filtered_force[:10]) 
    
    return filtered_height, filtered_force
```

## Contact Point Identification

```python
def get_noise_characterization(height, force, window_size):
    """
    Calculates the coefficient of determination (R^2) over a sliding window.

    This function moves a window across the force-indentation data and, for
    each position, calculates how well the data within that window fits a
    straight line. This is used to quantify the local linearity of the signal,
    which changes significantly at the point of contact where the cantilever
    engages the sample.

    Parameters
    ----------
    height : np.ndarray
        1D array of AFM piezo height values.
    force : np.ndarray
        1D array of AFM force values, corresponding to the height array.
    window_size : int
        The half-size of the rolling window used for the linear fit. The
        total window will be 2 * window_size points.

    Returns
    -------
    np.ndarray
        A 2D array where column 0 is the central height of the window and
        column 1 is the calculated R^2 value for that window.
    """
    r_array = []
    for i in range(window_size, height.size - window_size):
        subset_force = force[i - window_size:i + window_size]
        subset_height = height[i - window_size:i + window_size]
        # linear fit
        params = np.polyfit(subset_height, subset_force, 1)
        lin_force = np.polyval(params, subset_height)

        sum_residuals_squared = np.sum((subset_force - lin_force)**2)
        ss_squared = np.sum((subset_force - np.mean(subset_force))**2)
        r_array.append([height[i], 1 - sum_residuals_squared / ss_squared])

    return np.array(r_array)

def floating_statistics(x, data, window_size):
    """
    Calculates floating statistics (min, max, mean) over a sliding window.

    This is a general utility function that moves a window across a dataset
    and computes basic statistics for the data within that window.

    Parameters
    ----------
    x : np.ndarray
        1D array of the independent variable (e.g., time, height).
    data : np.ndarray
        1D array of the dependent variable to be analyzed.
    window_size : int
        The half-size of the rolling window.

    Returns
    -------
    np.ndarray
        A 2D array where the columns are:
        - 0: Central x value of the window.
        - 1: Minimum data value in the window.
        - 2: Maximum data value in the window.
        - 3: Mean data value in the window.
    """
    r_array = []
    for i in range(window_size, x.size - window_size):
        subset_data = data[i - window_size:i + window_size]
        # linear fit
        r_array.append([x[i], np.min(subset_data), np.max(subset_data), np.mean(subset_data)])

    return np.array(r_array)

def get_diff_start_x(x, data):
    """
    Identifies the contact point by maximizing the t-statistic between two populations.

    This function iterates through each point of the input data (typically the R^2
    array from get_noise_characterization) and treats it as a potential partition
    point. It splits the data into two samples (all points before and all points
    after) and calculates the t-statistic from an independent two-sample t-test.
    The point that maximizes this statistic corresponds to the most significant
    change in the signal's character, which is interpreted as the contact point.

    Parameters
    ----------
    x : np.ndarray
        1D array of the independent variable (e.g., height).
    data : np.ndarray
        1D array of the data whose transition point is to be found (e.g., R^2 values).

    Returns
    -------
    np.ndarray
        A 2D array where column 0 is the potential contact point (height) and
        column 1 is the calculated t-statistic. The peak of this array
        identifies the contact point.
    """
    r_array = []
    for i in range(1, x.size - 2):
        # Perform an independent two-sample t-test on the data before and after point i
        t_statistic = stats.ttest_ind(data[i:], data[:i])[0]
        r_array.append([x[i], t_statistic])
    return np.array(r_array)
    
# example data
%matplotlib inline
fig, ax = plt.subplots(nrows=2, ncols=2)
ax[0,0].plot(height, force, "-")

# Inset region definition
x1, x2 = 2.0e-6, 2.05e-6
y1, y2 = -0.01e-9, 0.015e-9
# Extract data for inset region
mask = (height >= x1) & (height <= x2)
h_inset = height[mask]
f_inset = force[mask]

# Inset axis
axins = ax[0,0].inset_axes([0.1, 0.6, 0.3, 0.3])  # [left, bottom, width, height]
axins.plot(h_inset, f_inset, ".")
axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)

x1, x2 = 7e-6, 7.05e-6
y1, y2 = 1.39e-9, 1.415e-9
# Draw box showing zoom region
ax[0,0].indicate_inset_zoom(axins, edgecolor="black")

mask = (height >= x1) & (height <= x2)
h_inset = height[mask]
f_inset = force[mask]

axins2 = ax[0,0].inset_axes([0.5, 0.6, 0.3, 0.3])  # [left, bottom, width, height]
axins2.plot(h_inset, f_inset, ".")
axins2.set_xlim(x1, x2)
axins2.set_ylim(y1, y2)
# Draw box showing zoom region
ax[0,0].indicate_inset_zoom(axins2, edgecolor="black")

ax[0,0].set_xlabel('height [m]')
ax[0,0].set_ylabel('force [N]')

window_size = 20
r_array = get_noise_characterization(height, force, window_size)

ax[0,1].plot(r_array[:,0], r_array[:,1], "b-", label = 'floating R^2')
ax[0,1].set_xlabel("height [m]")
ax[0,1].set_ylabel("floating R^2 [1]")



r_stat = floating_statistics(r_array[:,0], r_array[:,1], window_size)

r_xstart = get_diff_start_x(r_array[:,0], r_array[:,1])
ax[1,1].plot(r_xstart[:,0], r_xstart[:,1], "b-", label = 'diff')
ax[1,1].set_xlabel("height [m]")
ax[1,1].set_ylabel("diff sum R^2 [1]")

peaks, properties = find_peaks(r_xstart[:,1], prominence=10, width=5)


contact_h = r_xstart[peaks,0]
contact_f = np.interp(contact_h, height, force)
ax[1,0].plot(height, force, "-")
ax[1,0].plot(contact_h, contact_f, 'ro', markersize=10, label='Identified Contact Point')
ax[1,0].legend()
ax[1,0].set_xlabel('height [m]')
ax[1,0].set_ylabel('force [N]')

fig.savefig('Figure_CP.svg')

plt.show()
```

## Curve Alignment and Averaging

```python
%matplotlib agg

%matplotlib agg

def fit_jpk_file_noise(file_name: str):
    """
    Processes a single JPK force-spectroscopy file to find the contact point and normalize the data.

    This function implements a full pipeline for analyzing a raw AFM force-indentation
    curve. It identifies the contact point using a model-independent statistical
    method based on changes in the signal's noise characteristics. After identification,
    it normalizes the curve by shifting the origin to the contact point (0,0).

    The method involves:
    1. Reading the height and force data from a JPK file.
    2. Calculating the local linearity (R^2) of the signal in a sliding window.
    3. Performing a statistical test (t-test) to find the most significant transition point.
    4. Identifying the peak of the statistical curve as the contact point.
    5. Handling cases with multiple potential contact points by selecting the most
       physically plausible one based on height and force criteria.
    6. Normalizing the force and height data relative to the identified contact point.

    Parameters
    ----------
    file_name : str
        The filename of the JPK data file (e.g., 'ha_002.jpk-force') expected
        to be located in the './Data/' directory.

    Returns
    -------
    force : np.ndarray
        The normalized force data array, with the contact point at zero force.
    height : np.ndarray
        The normalized height (indentation) data array, with the contact point at zero height.
    """
    # Extract a unique identifier from the filename (e.g., 'ha_002').
    measurement_id = file_name.split('.')[0]

    # Construct the full path and read the AFM data using a helper function.
    path_file = "./Data/" + file_name
    height, force = read_jpk(path_file, 10e-9, 0e-6)

    # --- Contact Point Identification ---
    window_size = 20
    # Step 1: Quantify local signal linearity using a rolling R^2 calculation.
    r_array = get_noise_characterization(height, force, window_size)
    
    # Step 2: Calculate floating statistics on the R^2 signal.
    r_stat = floating_statistics(r_array[:,0], r_array[:,1], window_size)

    # Step 3: Use a t-test to find the most significant change in the signal.
    # This identifies the point of maximum difference between pre- and post-contact regions.
    r_xstart = get_diff_start_x(r_stat[:,0], r_stat[:,2])

    # Step 4: Find the peak(s) in the t-statistic curve. The highest peak corresponds to the contact point.
    peaks, properties = find_peaks(r_xstart[:,1], prominence=10, width=5)
    contact_h = r_xstart[peaks,0]
    contact_f = np.interp(contact_h, height, force)

    # --- Sanity Check for Multiple Detected Peaks ---
    # If more than one potential contact point is found, this logic selects the most
    # plausible one based on physical constraints (e.g., reasonable height and near-zero force).
    if contact_h.size > 1:
        # Define a plausible physical region for the contact point.
        idx_contact = (contact_h > 4e-6) & (contact_f < 0.2e-9)
        if not idx_contact.any():
            # If no peaks are in the plausible region, default to the first detected peak.
            contact_h = contact_h[0]
            contact_f = contact_f[0]
        else:
            # Otherwise, select the first peak that falls within the plausible region.
            contact_h = contact_h[idx_contact][0]
            contact_f = contact_f[idx_contact][0]

    # --- Verification Plot (Generated in Memory) ---
    # This plot is created for debugging or batch-saving purposes. The `plt.close(fig)`
    # command prevents the plot from being displayed when the script is run in a loop.
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.plot(height, force, 'k.', label='Raw Data')
    ax.plot(contact_h, contact_f, 'ro', markersize=10, label='Identified Contact Point')
    ax.set_xlabel("Height [m]")
    ax.set_ylabel("Force [N]")
    ax.set_title("Model Fit for Indentation Curve for " + measurement_id)
    ax.legend()
    ax.grid(True)
    plt.close(fig) # Prevents the figure from being displayed.
        
    # --- Data Normalization ---
    # Shift the entire curve so that the contact point is at the origin (0,0).
    # This transforms the raw data into a standardized force-indentation profile.
    force = force - contact_f
    height = height - contact_h
        
    return force, height



group = 'ha'
force_ha = []
height_ha = []

for file_name in file_names:
    if group in file_name:
        force, height = fit_jpk_file_noise(file_name)
        force_ha.append(force)
        height_ha.append(height)
        
group = 'pbs'
force_pbs = []
height_pbs = []

for file_name in file_names:
    if group in file_name:
        force, height = fit_jpk_file_noise(file_name)
        force_pbs.append(force)
        height_pbs.append(height)
```

## Force Deconvolution



```python
# --- Step 1: Find the maximum common indentation range across all curves ---

%matplotlib inline

fig, ax = plt.subplots(nrows=1, ncols=3, gridspec_kw={'width_ratios': [1,2,1]})

# Initialize a variable to track the shortest indentation depth for PBS liposomes.
h_max_pbs = np.inf
# Loop through all PBS (buffer-filled) curves.
for i in range(len(force_pbs)):
    # Plot each individual curve in blue.
    ax[0].plot(height_pbs[i], force_pbs[i], color="b")
    # Update the shortest maximum height. This ensures we don't try to average
    # beyond the range of the shortest measurement.
    h_max_pbs = np.min([h_max_pbs, np.max(height_pbs[i])])

# Repeat the process for HA (hyaluronic acid-filled) liposomes.
h_max_ha = np.inf
for i in range(len(force_ha)):
    # Plot each individual curve in red.
    ax[0].plot(height_ha[i], force_ha[i], color="r")
    h_max_ha = np.min([h_max_ha, np.max(height_ha[i])])

# The true common range is limited by the shortest curve across BOTH datasets.
h_min = np.min([h_max_ha, h_max_pbs])

# --- Step 2: Resample all curves onto a common axis and average them ---

# Create a new, uniform height axis with a 1 nm step size for interpolation.
h_common = np.arange(0, h_min, 1e-9)

# Process the PBS curves.
f_common_pbs = []
for idx in range(len(height_pbs)):
    # Use linear interpolation to resample the current force curve onto the common height axis.
    f_interp = np.interp(h_common, height_pbs[idx], force_pbs[idx])
    f_common_pbs.append(f_interp)

# Convert the list of resampled curves into a 2D NumPy array for calculation.
f_common_pbs = np.array(f_common_pbs)
# Calculate the mean force and standard deviation at each point along the common axis.
f_mean_pbs = np.mean(f_common_pbs, axis=0)
f_std_pbs =  np.std(f_common_pbs, axis=0)
h_common_pbs = h_common

# Process the HA curves using the same method.
f_common_ha = []
for idx in range(len(height_ha)):
    f_interp = np.interp(h_common, height_ha[idx], force_ha[idx])
    f_common_ha.append(f_interp)

f_common_ha = np.array(f_common_ha)
f_mean_ha = np.mean(f_common_ha, axis=0)
f_std_ha =  np.std(f_common_ha, axis=0)
h_common_ha = h_common

# --- Step 3: Isolate the Core's Contribution and Propagate Uncertainty ---

# Subtract the mean force of the membrane (PBS) from the mean force of the
# whole liposome (HA) to isolate the mechanical contribution of the viscous core.
f_diff = f_mean_ha - f_mean_pbs

# Calculate the combined standard deviation for the difference. This is done by
# adding the individual variances in quadrature (standard error propagation for subtraction).
std_diff = np.sqrt(f_std_ha**2 + f_std_pbs**2)
```

```python
%matplotlib inline

fig, ax = plt.subplots(nrows=1, ncols=3, gridspec_kw={'width_ratios': [1,2,1]})

ax[0].plot(h_common, f_mean_pbs, color="k", label="PBS liposome")
ax[0].plot(h_common, f_mean_pbs - f_std_pbs,"-.", color="k")
ax[0].plot(h_common, f_mean_pbs + f_std_pbs,"-.", color="k")

ax[0].plot(h_common, f_mean_ha, color="k", label="HA liposome")
ax[0].plot(h_common, f_mean_ha - f_std_ha,"-.", color="k")
ax[0].plot(h_common, f_mean_ha + f_std_ha,"-.", color="k")

ax[0].set_ylim([-0.5e-9,1.5e-9])
ax[0].set_xlim([-1e-6,1.5e-6])
ax[0].legend()

ax[1].plot(h_common_pbs, f_mean_pbs, color = 'b', label = "PBS liposome")
ax[1].fill_between(h_common_pbs, f_mean_pbs - f_std_pbs, f_mean_pbs + f_std_pbs, color = 'b', alpha=0.2)
ax[1].plot(h_common_ha, f_mean_ha, color= 'r', label = "HA liposome")
ax[1].fill_between(h_common_ha, f_mean_ha - f_std_ha, f_mean_ha + f_std_ha, color = 'r', alpha=0.2)
ax[1].plot(h_common_ha, f_diff, color = "g", label = "viscous core")
ax[1].fill_between(h_common_ha, f_diff - std_diff, f_diff + std_diff, color = 'g', alpha=0.2)
ax[1].set_ylim([-1e-10,0.1e-9])
ax[1].set_xlim([0,1e-7])
ax[1].legend()

hab = f_mean_pbs[h_common_pbs <= np.max(h_common_ha)]<f_diff
percentage_ha = 100*f_diff/f_mean_ha
err_ha = 100*std_diff/f_mean_ha
#err_ha = percentage_ha * np.sqrt((std_diff/f_diff)**2 +  (f_mean_ha/f_std_ha)**2)
percentage_pbs = 100*f_mean_pbs[h_common_pbs <= np.max(h_common_ha)]/f_mean_ha
err_pbs = 100*f_std_pbs[h_common_pbs <= np.max(h_common_ha)]/f_mean_ha
#err_pbs = percentage_pbs * np.sqrt((f_std_pbs[h_common_pbs <= np.max(h_common_ha)]/f_mean_pbs[h_common_pbs <= np.max(h_common_ha)])**2 +  (f_mean_ha/f_std_ha)**2)

ax[2].yaxis.set_major_formatter(mtick.PercentFormatter())
ax[2].plot(h_common_ha[hab], percentage_ha[hab]
           , color = "g",  label="% HA core")
ax[2].fill_between(h_common_ha[hab], percentage_ha[hab] - err_ha[hab], percentage_ha[hab] + err_ha[hab], color = 'g', alpha=0.2)
ax[2].plot(h_common_ha[hab], percentage_pbs[hab], color = "b",  label="% DPPC membrane")
ax[2].fill_between(h_common_ha[hab], percentage_pbs[hab] - err_pbs[hab], percentage_pbs[hab] + err_pbs[hab], color = "b",  alpha= 0.2)
ax[2].set_ylim([0,100])
ax[2].set_xlim([0,1e-6])
#data description
ax[2].legend()
fig.savefig('Figure_Force.svg')
plt.show()
```
